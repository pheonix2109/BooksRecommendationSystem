```python
ratings = pd.read_csv(r"C:\Users\palla\Downloads\Books\Ratings.csv")
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
x = ratings['User-ID'].value_counts() > 200    #divided the users who have rated the books more than 200 times and not in true and false
print(x)
```

```python
y = x[x].index     #taken the users-id of the users who have rated more than 200 books only
print(y)
```

```python
ratings = ratings[ratings['User-ID'].isin(y)]     #in ratings dataset now we have only the rows of users along with their ids who rated more than 200 books
```

```python
books_with_ratings = pd.merge(books, ratings, on = 'ISBN')     #merging dataframes rating with the books with column ISBN
```

```python
books_with_ratings.head()
```

```python
books_with_ratings.shape
```

```python
number_ratings = books_with_ratings.groupby('Book-Title')['Book-Rating'].count().reset_index()     #create a dataframe 'number_ratings', counted the number of ratings got by each book and stored
```

```python
number_ratings.rename(columns= {'Book-Rating':'Number-of-Ratings'}, inplace = True)     #renamed the columns 'Book-Title' to 'Number-of-Ratings'
print(number_ratings)
```

```python
final_books = pd.merge(books_with_ratings, number_ratings ,on = 'Book-Title')
```

```python
final_books.head()
```

```python
final_books = final_books[final_books['Number-of-Ratings'] >= 50]
```

```python
final_books.shape
```

```python
final_books.drop_duplicates(['Book-Title','User-ID'], inplace = True)
```

```python
final_books.shape
```

```python
book_pivot = final_books.pivot_table(columns = 'User-ID', index = 'Book-Title', values = 'Book-Rating')
```

```python
book_pivot.fillna(0, inplace = True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors= 6)
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```
